```python
#词向量
import pandas as pd
import jieba
from gensim.models.word2vec import Word2Vec

# 读入训练集文件
data = pd.read_csv('train.csv')
# 转字符串数组
corpus = data['comment'].values.astype(str)
# 分词，再重组为字符串数组
corpus = [jieba.lcut(corpus[index]
                          .replace("，", "")
                          .replace("!", "")
                          .replace("！", "")
                          .replace("。", "")
                          .replace("~", "")
                          .replace("；", "")
                          .replace("？", "")
                          .replace("?", "")
                          .replace("【", "")
                          .replace("】", "")
                          .replace("#", "")
                        ) for index in range(len(corpus))]
# 词向量模型训练
model = Word2Vec(corpus, sg=0, vector_size=300, window=5, min_count=3, workers=4)
#模型显示
print('模型参数：',model,'\n')
```

```python
print('最匹配的词是：',model.wv.most_similar(positive=['点赞', '不错'], negative=['难吃']),'\n')
```

```python
print('最不匹配的词是：',model.wv.doesnt_match("点赞 好吃 支持 难吃".split()),'\n')
```

```python
print('相似度为=',model.wv.similarity('推荐','好吃'),'\n')
```

```python
print(model.wv.__getitem__('地道'))
```

```python
model1 = Word2Vec(corpus, sg=1, vector_size=300, window=5, min_count=3, workers=4)
print('新模型参数：',model1,'\n')
```

```python
env_vector =model1.wv.__getitem__('环境')
print("词向量：", env_vector)
print("形状：", env_vector.shape)
```

```python
similar_words = model1.wv.most_similar('好吃', topn=3)
print("最接近的3个词：", similar_words)
```

```python
sim1 = model1.wv.similarity('好吃', '美味')
sim2 = model1.wv.similarity('好吃', '蟑螂')
print("好吃 vs 美味：", sim1)
print("好吃 vs 蟑螂：", sim2)
```

```python
result = model1.wv.most_similar(positive=['餐厅', '聚会'], negative=['安静'], topn=1)
print("最相关结果：", result[0])
```
